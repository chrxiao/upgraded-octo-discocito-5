```python
import time
from gensim.models.keyedvectors import KeyedVectors

path = "data/glove.6B.50d.txt.w2v"
t0 = time.time()
glove = KeyedVectors.load_word2vec_format(path, binary=False)
t1 = time.time()
print("elapsed %ss" % (t1 - t0))
```

```python
with open("data/stopwords.txt", 'r') as r:
    stops = []
    for line in r:
        stops += [i.strip() for i in line.split('\t')]
```

```python
word = "scrumptious"

if word in glove.vocab:
    print(glove[word])
```

```python
captions = ["The cat in the hat.", "Sam is very wise (and very strong)", "The cat jumped over Sam", "$#!^cat@@$*&&@)$&@"]

reduce_captions(captions, glove, stops)
```
